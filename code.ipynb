```python
import json
import pandas as pd

with open("data/null.json") as f:
    data = json.load(f)

null_df = pd.DataFrame.from_dict(data, orient="index")
null_df.reset_index(inplace=True)
null_df.columns = ["index", "value1", "value2", "value3"][:len(null_df.columns)]
null_df["index"] = null_df["index"].str.upper()
# Replace string "None" with np.nan in value1 and value2
null_df['value1'] = null_df['value1'].replace("None", np.nan)
null_df['value2'] = null_df['value2'].replace("None", np.nan)
null_df['value3'] = null_df['value3'].replace("None", np.nan)

# Convert value1 and value2 columns to numeric types, coercing errors to NaN
null_df['value1'] = pd.to_numeric(null_df['value1'], errors='coerce')
null_df['value2'] = pd.to_numeric(null_df['value2'], errors='coerce')
null_df['value3'] = pd.to_numeric(null_df['value3'], errors='coerce')
null_df = null_df[~null_df["index"].isin(cols_to_drop)].reset_index(drop=True)
null_df
```

```python
def replace_with_nulls(data_df, null_df):
    """
    Replace specified values in data_df with NaN based on null_df.
    """
    result_df = data_df.copy()

    for _, row in null_df.iterrows():
        col = row.get("variable", row.get("index"))
        if col not in result_df.columns:
            print(f"Warning: Column '{col}' not in data_df. Skipping.")
            continue

        # Get dtype of the column
        col_dtype = result_df[col].dtype

        # Prepare values to replace with correct type
        vals_to_replace = []
        for key in ["value1", "value2", "value3"]:
            val = row.get(key)
            if pd.notna(val):
                try:
                    val = col_dtype.type(val)  # cast to same type as column
                    vals_to_replace.append(val)
                except Exception as e:
                    print(f"Could not cast value {val} for column {col}: {e}")

        if vals_to_replace:
            result_df[col] = result_df[col].replace(vals_to_replace, pd.NA)

    return result_df

data_df_na = replace_with_nulls(data_df, null_df)
```
