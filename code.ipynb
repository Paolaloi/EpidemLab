```python
import pandas as pd
import numpy as np

def replace_with_nulls(data_df, null_df):
    """
    Replace specific values in a dataframe with NaN based on specifications in null_df.
    
    Parameters:
    -----------
    data_df : pandas.DataFrame
        The dataframe containing the data to process
    null_df : pandas.DataFrame
        A dataframe with three columns:
        - 'variable': column names in data_df
        - 'value1': primary values to replace with NaN
        - 'value2': optional secondary values to replace with NaN
    
    Returns:
    --------
    pandas.DataFrame
        A copy of data_df with specified values replaced by NaN
    """
    # Create a copy of the dataframe to avoid modifying the original
    result_df = data_df.copy()
    
    # Iterate through each row in null_df
    for idx, row in null_df.iterrows():
        column_name = row['variable']
        
        # Check if the column exists in data_df
        if column_name not in result_df.columns:
            print(f"Warning: Column '{column_name}' specified in null_df not found in data_df. Skipping.")
            continue
        
        # Replace value1 with NaN if it's not null/NaN
        if pd.notna(row['value1']):
            result_df.loc[result_df[column_name] == row['value1'], column_name] = np.nan
        
        # Replace value2 with NaN if it's not null/NaN
        if 'value2' in row.index and pd.notna(row['value2']):
            result_df.loc[result_df[column_name] == row['value2'], column_name] = np.nan
    
    return result_df

data_df_na = replace_with_nulls(data_df, null_df)
```
