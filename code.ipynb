```python
import json

with open("data/null.json") as f:
    data = json.load(f)

null_df = pd.DataFrame.from_dict(data, orient="index")
null_df.reset_index(inplace=True)
null_df.columns = ["index", "value1", "value2", "value3"][:len(null_df.columns)]
null_df["index"] = null_df["index"].str.upper()
# Replace string "None" with np.nan in value1 and value2
null_df['value1'] = null_df['value1'].replace("None", np.nan)
null_df['value2'] = null_df['value2'].replace("None", np.nan)
null_df['value3'] = null_df['value3'].replace("None", np.nan)

# Convert value1 and value2 columns to numeric types, coercing errors to NaN
null_df['value1'] = pd.to_numeric(null_df['value1'], errors='coerce')
null_df['value2'] = pd.to_numeric(null_df['value2'], errors='coerce')
null_df['value3'] = pd.to_numeric(null_df['value3'], errors='coerce')
null_df = null_df[~null_df["index"].isin(cols_to_drop)].reset_index(drop=True)
null_df
```

```python
def replace_with_nulls(data_df, null_df):
    """
    Replace specified values in data_df with NaN based on null_df.
    """
    result_df = data_df.copy()

    for _, row in null_df.iterrows():
        col = row.get("variable", row.get("index"))
        if col not in result_df.columns:
            print(f"Warning: Column '{col}' not in data_df. Skipping.")
            continue

        # Get dtype of the column
        col_dtype = result_df[col].dtype

        # Prepare values to replace with correct type
        vals_to_replace = []
        for key in ["value1", "value2", "value3"]:
            val = row.get(key)
            if pd.notna(val):
                try:
                    val = col_dtype.type(val)  # cast to same type as column
                    vals_to_replace.append(val)
                except Exception as e:
                    print(f"Could not cast value {val} for column {col}: {e}")

        if vals_to_replace:
            result_df[col] = result_df[col].replace(vals_to_replace, pd.NA)

    return result_df

data_df_na = replace_with_nulls(data_df, null_df)
```

```python
data_df_encoded['ALC_TOTAL'] = data_df_encoded[['ALC4', 'ALC8', 'ALC12', 'ALC16', 'ALC20', 'ALC25']].sum(axis=1, skipna=True)
data_df_encoded['FUM_TYPE1_YEARS'] = data_df_encoded[['FUM9', 'FUM8']].assign(FUM8=lambda x: -x['FUM8']).sum(axis=1, skipna=True)
data_df_encoded['PRE_BMI'] = data_df_encoded['ANTR1'] / (data_df_encoded['ANTR2'] / 100) ** 2 # Body Mass Index (BMI) pre diagnosis
data_df_encoded['WHR'] = data_df_encoded['ANTR3'] / data_df_encoded['ANTR4'] # Waist-to-Hip Ratio (WHR)
#data_df_encoded['WEIGHT_MAX_DIFF'] = data_df_encoded['ANTR8'] - data_df_encoded['ANTR10'] # Max weight variation: max - min since 20 years
data_df_encoded['AVG_USUAL_WEIGHT'] = data_df_encoded[['ANTR6', 'ANTR7']].mean(axis=1, skipna=True) # Usual Weight 30-50 years old
```

```python
pd.set_option('future.no_silent_downcasting', True)
corr_matrix = data_df_encoded.fillna(np.nan).infer_objects(copy=False).corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find columns with correlation above threshold (e.g., 0.9)
threshold = 0.9
high_corr = [(col, row, upper.loc[row, col])
             for col in upper.columns
             for row in upper.index
             if upper.loc[row, col] > threshold]

# Convert to DataFrame for easier viewing
high_corr_df = pd.DataFrame(high_corr, columns=["Variable 1", "Variable 2", "Correlation"])
```

```python
high_corr_df
```

```python
import networkx as nx

G = nx.Graph()
for _, row in high_corr_df.iterrows():
    G.add_edge(row["Variable 1"], row["Variable 2"])

removable_vars = set()
for component in nx.connected_components(G):
    # For each group, count the number of high correlations for each variable
    subgraph = G.subgraph(component)
    degrees = dict(subgraph.degree())
    # Keep the variable with the highest degree (most connections)
    keep_var = max(degrees, key=degrees.get)
    # Mark all others as removable
    removable_vars.update(set(component) - {keep_var})

removable_vars = sorted(removable_vars)
print("Variables to consider removing:", removable_vars)
```
